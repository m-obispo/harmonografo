```python
# Importando bibliotecas
from matplotlib.animation import PillowWriter
import matplotlib.animation as animation
from IPython.display import HTML, Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
# Definindo a função pendulo com valor de parâmetros padrão
def Pendulo(t, A = 1, f = 4, p = 0, d = 0.05):
    return (A*np.sin(2*np.pi*f*t + p)*np.exp(-d*t))
```

```python
# Criando um array para t
t = np.linspace(0, 50*np.pi, 30000)
```

```python
# Armazenando o tempo em matrizes
xt = np.zeros(len(t))
yt = np.zeros(len(t))
```

```python
# Plotando os resultados
fig = plt.figure(figsize = (14, 9), dpi = 65)
ax = fig.add_subplot()
ax.set_facecolor('k')
plt.style.use(['dark_background'])
#plt.close(fig)

# Criando a animação
def update(i):
    xt[i] = Pendulo(t[i], 2, 4, -1, 0.05) + Pendulo(t[i], 1, 2, np.pi, 0.01)
    yt[i] = Pendulo(t[i], 4, 2, np.pi/4, 0.02) + Pendulo(t[i], 1, 2, np.pi/2, 0.25) 
    ax.clear()
    ax.plot(xt[:i], yt[:i], 'g')
    #ax.grid(color = 'b', linestyle = '-', linewidth = 1)
    ax.set_title('$Harmonógrafo$: t = ' + str(i), color = 'g', fontsize = 18)
    ax.set_xlabel('$x(t)$', color = 'g', fontsize = 15)
    ax.set_ylabel('$y(t)$', color = 'g', fontsize = 15)

# Chama a função update() periodicamente
ani = animation.FuncAnimation(fig, update, np.arange(5000), interval = 10, repeat = False)
#ani.save('2eD.gif', writer = PillowWriter(fps = 100)) #Harmonógrafo2D
#Image(url = '2eD.gif')
plt.show()
```
